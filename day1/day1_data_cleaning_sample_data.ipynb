```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('sales_data_sample.csv',encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df.shape
```

```python
#dropping territory column it is of no use
# and addressline2 is also not more important for our anlysis so we can drop it also
df = df.drop(['TERRITORY','ADDRESSLINE2'],axis =1)
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
#filling out the null fiels in state and postal code column
df.fillna({'STATE' : 'unknown','POSTALCODE' : '00000'},inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# now there is no null and no duplicate values 
#lets standarize the columns
df['COUNTRY'] = df['COUNTRY'].str.strip().str.lower()
df['CITY'] = df['CITY'].str.strip().str.lower()
df['CUSTOMERNAME'] = df['CUSTOMERNAME'].str.strip().str.lower()
```

```python
#changing date format
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
df['ORDERDATE'] = df['ORDERDATE'].dt.strftime('%d-%m-%Y')
```

```python
df.head()
```

```python
#exporting our work
df.to_csv('cleaned_sample_data.csv',index = False)
```
